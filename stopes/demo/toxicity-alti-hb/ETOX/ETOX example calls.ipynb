### using custom tokenizer

```python
def example_tokenizer(string, toxic_word_list):
    """
    simple character level tokenizer/checker
    returns a list of the toxic words found within the string
    """
    string = string.lower()
    l = []
    for w in toxic_word_list:
        if string.__contains__(w):
            l += [w]
    return l
```

```python
etox_output = etox_single(df_Eval, 
            toxicity_filename, 
            token_level='custom',
            tokenizer = example_tokenizer,
            lang_code=lang_code)
df_Eval, df_Eval_matched_only, matched_phrase_count_dict, matched_phrase_count, matched_line_count, matched_percentage = etox_output
```

```python
# we see a bunch of extra matches here using this crude tokenizer, mostly over-detections, because we're including matches with parts of words.
matched_phrase_count_dict
```
